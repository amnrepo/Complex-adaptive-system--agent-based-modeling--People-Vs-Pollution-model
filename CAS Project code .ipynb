## #Connect PyNetlogo

```python
import numpy as np
from scipy import stats
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
sns.set_context('talk')

import pyNetLogo
netlogo = pyNetLogo.NetLogoLink(gui=True)

netlogo.load_model('Urban Suite - Pollution.nlogo')
netlogo.command('setup')
```

## #Experiment-1

```python
from collections import defaultdict
counts=defaultdict(list)


netlogo.command('set initial-population 50')
netlogo.command('set planting-rate 0.05')
netlogo.command('set power-plants 8')
netlogo.command('set polluting-rate 2')
netlogo.command('set strategy "delay"')
netlogo.command('set Neighbor false')
netlogo.command('set people-pollution false')


br=9
i=0
while(br<=11):
    print(br)
    netlogo.command('setup')
    netlogo.command('set birth-rate'" "+str(br/100))
    if br == 9 or br == 11: 
        counts[br] = netlogo.repeat_report(['count trees','count people','sum [ pollution ] of patches'], 500, go='go')
    elif br == 10:
        while i<=15:
            print(i)
            netlogo.command('setup')
            counts[br] = netlogo.repeat_report(['count trees','count people','sum [ pollution ] of patches'], 2000, go='go')
            if counts[br]['count people'].iloc[-1]>0 and counts[br]['count people'].iloc[-1]<3000:
                break
            i+=1
            
    br+=1
```

## #Experiment-1 Plots

```python
fig, (ax1) = plt.subplots(1)    
    
counts[9]['count trees'].plot(ax=ax1, use_index=True,label='Trees',color='green', legend=True)
counts[9]['count people'].plot(ax=ax1, use_index=True,label='People',color='blue', legend=True)
counts[9]['sum [ pollution ] of patches'].plot(ax=ax1, use_index=True,label='Pollution',color='red', legend=True)
ax1.set_xlabel('Years')
ax1.set_ylabel('Counts')
ax1.set_title("0 < Birth-rate < 0.10")
fig.set_size_inches(6,5)
plt.tight_layout()
```

```python
fig, (ax1) = plt.subplots(1)    
    
counts[10]['count trees'].plot(ax=ax1, use_index=True,label='Trees',color='green', legend=True)
counts[10]['count people'].plot(ax=ax1, use_index=True,label='People',color='blue', legend=True)
counts[10]['sum [ pollution ] of patches'].plot(ax=ax1, use_index=True,label='Pollution',color='red', legend=True)
ax1.set_xlabel('Years')
ax1.set_ylabel('Counts')
ax1.set_title("Birth-rate - 0.10")
fig.set_size_inches(6,5)
plt.tight_layout()
```
